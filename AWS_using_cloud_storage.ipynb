```python
# Create a aws client to connecto to s3
client = boto3.client('s3')
```

```python

# Get the list of objects in s3
response = client.list_objects(
    Bucket = "kelp-satellite-data"
)
```

```python
# iterate over names
for i in response.get("Contents",None):
    print(i.get("Key",None))
# use .get("key",None) because if the keys its not found, the code will not break down
# instead it return us "None"
```
